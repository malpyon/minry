```python
import pandas as pd
from scipy import stats
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('tugas.csv', sep=";")
df.head()
```

```python
df.describe()
```

```python
data= {"Statistics": ['Count', 'Max', 'Min', 'Mean','Standard Deviasi', 'Variasi', 'Quantile 1', 'Quantile 2',
                      'Quantile 3', 'Skewnes', 'Median', 'Modus']}
for i in df.columns:
    data[i] = [df[i].count(), df[i].max(), df[i].min(), df[i].mean(), round(df[i].std(),2), round(df[i].var(),2),
               df[i].quantile(0.25), df[i].quantile(0.5), df[i].quantile(0.75), round(df[i].skew(),2), df[i].median(),
              stats.mode(df[i]).mode[0]]
tes = pd.DataFrame(data,columns = ['Statistics'] + [x for x in df.columns])
tes
```
